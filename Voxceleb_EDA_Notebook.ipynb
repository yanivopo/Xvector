### Import Dependencies

```
import pandas as pd
import numpy as np
%pylab notebook 
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import pandas_profiling
sns.set_style('whitegrid')
```

### Read CSV file

```
df = pd.read_csv('vox1_meta.csv', sep = '\t')
```

```
df.head()
```

## Show some insights

```
fig, ax = plt.subplots(figsize = (10,8))
ax.pie(df.Gender.value_counts(), labels = ['Male', 'Female'],
      autopct='%1.1f%%', textprops={'fontsize': 14})
ax.set_title('Male vs Female', fontsize=20) 
fig.tight_layout()
plt.show()
```

```
labels = [round(i / len(df) * 100,1) for i in df.Nationality.value_counts()[:10]]
```

```
fig, ax = plt.subplots(figsize = (10,12))
ax.pie(df.Nationality.value_counts()[:10], 
      labels = labels, textprops={'fontsize': 14})
ax.set_title('Top 10 Speekers Nationality', fontsize=20) 
fig.tight_layout()
plt.legend(df.Nationality.value_counts().index.to_list()[:10])
plt.show()
```

```
plt.figure(figsize = (20,10))
sns.countplot(y = 'Nationality', hue = 'Gender', data = df)
plt.title('Gender counts per nationality', fontsize = 15)
plt.show()
```

```
plt.figure(figsize = (12,10))
sns.countplot(x = 'Gender', hue = 'Set', data = df, palette = 'Set1')
plt.title('Gender counts per train/val set', fontsize = 15)
plt.show()
```
